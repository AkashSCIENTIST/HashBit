# CSV to JSON

```python
import csv, json
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./sample_data.csv")
```

```python
df.columns
```

```python
df.rename(columns={**{k : k.lower().strip() for k in df.columns}, **{'Unnamed: 3': 'description'}}, inplace=True)
```

```python
df["description"].fillna("", inplace=True)
```

```python
df
```

```python
df.shape
```

```python
df.to_csv("newdata.csv", index=False)
```

```python
data_list = list(df.T.to_dict().values())
```

```python
data_string = json.dumps(data_list, indent=2)
```

```python
with open("data.json", "w") as f:
    f.write(data_string)
```
